# 3.**Data Cleaning**

### 3.1.check Nan values in columns

```python
df.isnull().sum() * 100 / len(df)
```

### 3.2.Drop Nan values in columns and rows and fill with zero

- Because columns `Notes` and `Status` have all values are Nan, Then will be drop it


```python
clean_df= df.drop(columns=['Notes', 'Status'])
clean_df.JobTitle.replace('', np.nan, inplace=True)
clean_df.Benefits=clean_df.Benefits.fillna(0)
clean_df= clean_df.dropna()
clean_df
```

```python
clean_df.isnull().sum()
```

### 3.3.Drop nagitve values in columns



```python
clean_df=clean_df[(clean_df["BasePay"] > 0) &( clean_df["OvertimePay"] > 0) & (clean_df["TotalPay"] > 0)]
clean_df.describe()
```

### 3.4.Convert **JobTitle** to LowerCase



```python
print(f"job Title before convert is: {clean_df.JobTitle.nunique()}")
clean_df.JobTitle=clean_df.JobTitle.str.lower().str.strip().replace(' ', '', regex=True)
print(f"job Title after convert is: {clean_df.JobTitle.nunique()}")
```

```python
print(f"data after clean is {clean_df.shape[0]/df.shape[0]} of orginal")
```

# 5.**Grouped Analysis**


```python
def convert(x):
  if(x<5000):
    return "NO_OVER_TIME"
  if(x<10000):
    return "FEW"
  return "OVER"


clean_df["OvertimePayStatus"]= clean_df.OvertimePay.map(convert)
```
